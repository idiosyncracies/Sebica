```python
path = '/home/royliu/Documents/datasets/Vehicle_Classification_kaggle'
import os
from PIL import Image

def get_sorted_image_sizes(path):
    # List to store sizes (width, height)
    image_sizes = []

    # Walk through all subdirectories and files
    for root, dirs, files in os.walk(path):
        for filename in files:
            if filename.endswith('.jpg'):
                # Open the image
                img_path = os.path.join(root, filename)
                with Image.open(img_path) as img:
                    # Get image size (width, height)
                    size = img.size
                    # Add size to the list
                    image_sizes.append(size)

    # Sort the sizes by width first, then by height
    sorted_sizes = sorted(image_sizes, key=lambda x: (x[0], x[1]))

    return sorted_sizes

# Example usage

sorted_sizes = get_sorted_image_sizes(path)
print("Sorted image sizes in the folder and subfolders:", sorted_sizes)
```

```python
import os
from PIL import Image

def resize_and_save_images(path):
    # Define output directories for high resolution and low resolution images
    hr_dir = os.path.join(path,'./all/HR')
    lr_dir = os.path.join(path,'./all/LR')
    
    # Create directories if they don't exist
    os.makedirs(hr_dir, exist_ok=True)
    os.makedirs(lr_dir, exist_ok=True)

    # Walk through all subdirectories and files
    for root, dirs, files in os.walk(path):

        for filename in files:
            if filename.endswith('.jpg'):
                # Open the image
                img_path = os.path.join(root, filename)
                with Image.open(img_path) as img:
                    # Convert to RGB if the image is in a mode not supported by JPEG
                    if img.mode != 'RGB':
                        img = img.convert('RGB')
                        
                    # Resize image to 512x512 and save to HR folder with best quality
                    hr_img = img.resize((640, 360))
                    hr_img.save(os.path.join(hr_dir, filename), quality=95)  # Best quality compression

                    # Resize image to 128x128 and save to LR folder with default settings
                    lr_img = img.resize((160, 90))
                    lr_img.save(os.path.join(lr_dir, filename))  # Default quality

                    
# Example usage
path = '/home/royliu/Documents/datasets/Vehicle_Classification_kaggle'
resize_and_save_images(path)
print("Images resized and saved to './HR' (512x512 with best quality) and './LR' (128x128 with default quality) folders.")
```
