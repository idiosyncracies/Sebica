## Copy lables to Traffic_Aerial_Images_for_Vehicle_Detection/val_labels
- Copy all val_folders' labels into xx/val_labels
- Manually seperate them by sec2, sec6, sec9...

```python

import os
import shutil
import os
root = '/home/royliu/Documents/datasets/Traffic_Aerial_Images_for_Vehicle_Detection'
val_folders = ['sec2','sec6','sec9']
# train_folders =  ['sec2','sec3','sec4', 'sec8', 'secb']
target_folder = os.path.join(root, 'val_labels')

# Create the target folder if it doesn't exist
if not os.path.exists(target_folder):
    os.makedirs(target_folder)

# Loop through each validation folder and search for text files
for folder in val_folders:
    folder_path = os.path.join(root, folder)
    for subdir, _, files in os.walk(folder_path):
        for file in files:
            if file.endswith('.txt'):
                file_path = os.path.join(subdir, file)
                # Copy the text file to the target folder
                shutil.copy(file_path, target_folder)
                print(f'Copied {file_path} to {target_folder}')

```
